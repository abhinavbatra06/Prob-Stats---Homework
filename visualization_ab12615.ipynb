```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
import plotly.graph_objs as go
```

```python

# generate error 1 

a_tilda = np.random.normal(0, 5, 500)
b_tilda = np.random.normal(0, 2, 500)
c_tilda = np.random.normal(0, 3, 500)

# generate x values

x = a_tilda + c_tilda

# generate y values
y = a_tilda + b_tilda

# generate z values
z = b_tilda - c_tilda



# creating a dataframe : 
df = pd.DataFrame({'a' : a_tilda , 'b' : b_tilda , 'c' : c_tilda , 'x': x, 'y': y, 'z': z})

# dumpingt to csv
df.to_csv('variables_for_viz_ab12615.csv', index = False)
```

```python
print("Correlation between x and y: ", np.corrcoef(x, y)[0, 1])
print("Correlation between y and z: ", np.corrcoef(y, z)[0, 1])
print("Correlation between x and z: ", np.corrcoef(x, z)[0, 1])

# Create a DataFrame with the three variables
data = {'x': x, 'y': y, 'z': z}
df_tilda = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df_tilda.corr()
print("\nCorrelation matrix:\n", correlation_matrix)
```

```python
import numpy as np
import plotly.graph_objects as go

X = x
Y = y
Z = z

# Create a 3D scatter plot using Plotly
fig = go.Figure()

# X-Y Relationship
fig.add_trace(go.Scatter3d(
    x=X, y=Y, z=np.zeros_like(X) - 10,  # Shifting along Z-axis to avoid overlap
    mode='markers',
    marker=dict(size=3, color=X, colorscale='Blues', opacity=0.8),
    name='X-Y Relation'
))

# Y-Z Relationship
fig.add_trace(go.Scatter3d(
    x=np.zeros_like(Y) + 10, y=Y, z=Z,  # Shifting along X-axis
    mode='markers',
    marker=dict(size=3, color=Y, colorscale='Greens', opacity=0.8),
    name='Y-Z Relation'
))

# Z-X Relationship
fig.add_trace(go.Scatter3d(
    x=X, y=np.zeros_like(Z) - 10, z=Z,  # Shifting along Y-axis
    mode='markers',
    marker=dict(size=3, color=Z, colorscale='Reds', opacity=0.8),
    name='Z-X Relation'
))

# Improve Layout
fig.update_layout(
    title="3D Scatter Plot for X-Y, Y-Z, and Z-X Relationships",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    margin=dict(l=0, r=0, b=0, t=50)
)

# Show interactive plot
fig.show()
```
